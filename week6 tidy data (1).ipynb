# data 2
https://data.ny.gov/Energy-Environment/Gasoline-Retail-Prices-Weekly-Average-by-Region-Be/nqur-w4p7/data


```python
df2 = pandas.read_csv("C:/Users/pragathi/Downloads/Gasoline_Retail_Prices_Weekly_Average_by_Region__Beginning_2007.csv")
df2.shape
```

```python
df2.head()
```

```python
tidy_df2 = pandas.melt(df2,
                       ["Date"],
                       var_name="city name",
                       value_name="average price($/gal)")
tidy_df2 = tidy_df2.sort_values(by=["city name"])
tidy_df2.head()
```

```python
tidy_df2.shape
```

```python
for column in tidy_df2.columns:
    print(column)
```

### Is this data tidy or not?

(no)

```python
id_vars=["Full Certificate Code","Certificate Type Code","Certificate Type Description","Authority Type Code","Authority Type Description","Business Type Code","Business Type Description","Name Type Code","Name Type Description","Description","Year","Total"]
tidy_df3 = pandas.melt(frame=df3,id_vars=id_vars, var_name="month", value_name="no.of filings")
```

```python
tidy_df3.head()
```

```python
for column in tidy_df3.columns:
    print(column)
```
