```python
from faker import Faker
import pandas as P
import numpy as N
import random
```

```python
fake=Faker()
```

```python
def fakedata(num):
    ownership=['rent','own']
    data=[{'email':fake.email(),
          'phonenumber':random.randint(1000000000,9999999999),
          'homeaddress':fake.address(),
          'personname':fake.name(),
          'yearborn':random.randint(1950,2000),
          'no.of kids':random.randint(1,4),
          'categorical variable':N.random.choice(ownership),
          'annual income':random.randint(100000,9999999),
          'speeding tickets':random.randint(1,5)}for i in range(num)]
    data_df = P.DataFrame(data)
    data_df.to_csv('fakedata week3.csv')
    return data_df
```

```python
fakedata(15)
```
